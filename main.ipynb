```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Loan.csv")
```

```python
data.head()
```

EDA

Check for missing values

```python
plt.figure(figsize=(30,6))
plt.subplot(1,3,1)
plt.title("Dataset Null")
sns.heatmap(data.isnull())
plt.show()
```

```python
null_count = data.isnull().sum()
print(null_count)
```

plotting the features

```python
employment_count = data["EmploymentStatus"].value_counts()

#plotting the bar chart
plt.figure(figsize=(7,4))
sns.barplot(x=employment_count.index, y=employment_count.values)
plt.title("Employment status")
plt.xlabel("Employment Type")
plt.ylabel("Number of employees")
plt.xticks(rotation=90)
plt.show()
```

```python
employment_count = data["EducationLevel"].value_counts()

#plotting the bar chart
plt.figure(figsize=(7,4))
sns.barplot(x=employment_count.index, y=employment_count.values)
plt.title("Education of employees")
plt.xlabel("Education Rank")
plt.ylabel("Number of employees")
plt.xticks(rotation=90)
plt.show()
```

```python
employment_count = data["MaritalStatus"].value_counts()

#plotting the bar chart
plt.figure(figsize=(7,4))
sns.barplot(x=employment_count.index, y=employment_count.values)
plt.title("Marital status of employees")
plt.xlabel("Status")
plt.ylabel("Number of employees")
plt.xticks(rotation=90)
plt.show()
```
